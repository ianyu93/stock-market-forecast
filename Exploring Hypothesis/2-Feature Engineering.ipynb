## **Feature Engineering**

There are 9,021 rows with 13 featues, and we will create new features to measure momentum, cross-market relative strength, and volitility.

#### Momentum

To measure momentum, we will use [Exponential Moving Average](https://www.investopedia.com/terms/e/ema.asp), or EMA. While Simple Moving Average (SMA, the regular moving average) smooths out the overall trend line, EMA adds more weight on more recent values. In other words, EMA not only shows the overall trend, it also incorporates latent momentum. We will be applying EMA on other features to learn momentum of other features.

*Note: We will be using pandas [Exponential Weighted Functions](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html), which works slightly different from the traditional EMA in the finance world, but serves our purpose, and inter*

#### Cross-Market Relative Strength

To measure cross-market relative strength, we will be using Relative Strength Line. A concept frequently used by John Murphy in Intermarket Analysis, the RSL is simply the ratio between any two given assets/markets. It measures the relative performance of the two assets. For example, SPX/US10Y RSL measures the relative performance of the stock market comparing to the 10 Year Treasury Bond. An increase of this indicator would signal that the stock market's relative performance to the bond market is getting stronger.

#### Volatility

To measure volatility, we will be using the [Bollinger Band](https://www.investopedia.com/terms/b/bollingerbands.asp) indicator. The Bollinger Band is comprised of a middle band of SMA, typically a 20-Day Moving Average, an upper band of +2 standard deviation from the middle band, and -2 standard deviation from the middle band. Approximately 90% of the price action would happen within the bands, and since standard deviation is a measure of volitility, when the market becomes volitile, the bands would expand, and vice versa when the market becomes more stable. 

#### Timeframe

For each of the ascribed feature, we will aslo be incorporating certain timeframe. For moving averages, we will use 5 (trading days) to measure weekly moving average, 20 (4 weeks of trading days) to represent monthly, and 60 (3 months) to represent a quarter. We will also employ market convention of 50-day and 200-day moving averages to measure long-term trends, despite the fact that they do not represent an intuitive timeframe. Contrary to popular belief, the heavy focus of the quantitative perspective in the financial world is actually relatively new (starting around the 80s). Many of the convention are just hard-tested rules that lasted for decades.

First of all, we want to create multiple RSL that compares the stock market against the dollar, the bond market, the oil market, and the gold market, creating RSL for `SPX/DXY`, `SPX/WTI`, `SPX/US10Y`, and `SPX/GOLD` respectively. Additionally, as both gold and bond markets are unique hedging asset, tied to the monetary environment, we will also create a RSL that measures their relative performance to each other. All five of these are going to be daily values.

```python
### Adding features
## Daily featuers
# SPX/DXY Relative Strength Line, increase means stock market is performing better
df['SPX/DXY RSL'] = df['SPX Close'] / df['DXY Close']

# SPX/WTI Relative Strength Line, increase means stock market is performing better
df['SPX/WTI RSL'] = df['SPX Close'] / df['WTI Price']

# SPX/US10Y Relative Strength Line, increase means stock market is performing better
df['SPX/US10Y RSL'] = df['SPX Close'] / df['US10Y Rate']

# SPX/GOLD Relative Strength Line, increase means stock market is performing better
df['SPX/GOLD RSL'] = df['SPX Close'] / df['GOLD Price']

# GOLD/US10Y Relative Strength Line, increase means gold market is performing better
df['GOLD/US10Y RSL'] = df['GOLD Price'] / df['US10Y Rate']
```

Next, we will create features for weekly moving average. Not every asset has to be tracked in every timeframe, and we are making such decision based on conventional understanding of each market's volitility. First of all, we will create an EMA5 for S&P 500 alone. Then we will track the EMA for stock market's performance against the currency market and the oil market, for these two are relatively more volitile than other markets. By measuring the EMA of RSL, we can capture the momentum if the stock market is increasingly stronger or weaker against a specific market. 

We will also create a simple moving average for `WTI/GOLD RSL` pair. While both oil and gold markets are a proxy to the commodity market, this feature can copture the dynamic within the commodity market at a very course level on a weekly basis. We will also include a `WTI/DXY` pair to capture the momentum of the dynamic bewteen the dollar and the oil market using EMA.

```python
## Features for weekly moving average
# SPX, EMA 5 to capture momentum
df['SPX EMA5'] = df['SPX Close'].ewm(span=5).mean()

# SPX/DXY Relative Strength Line, 5-day simple moving average to capture general trend 
df['SPX/DXY RSL MA5'] = df['SPX/DXY RSL'].rolling(window=5).mean()

# SPX/WTI Relative Strength Line, 5-day simple moving average to capture general trend 
df['SPX/WTI RSL MA5'] = df['SPX/WTI RSL'].rolling(window=5).mean()

# WTI/GOLD Relative Strength Line, 5-day simple moving average to capture general trend 
df['WTI/GOLD RSL MA5'] = (df['WTI Price'] / df['GOLD Price']).rolling(window=5).mean()

# WTI/DXY Relative Strength Line, exponential moving average to capture momentum
df['WTI/DXY RSL EMA5'] = (df['WTI Price'] / df['DXY Close']).ewm(span=5).mean()
```

Next up we will be creating features for a timeframe of one month. We will create a EMA13 and EMA 26 for S&P 500 as convention goes. We will also measure a 20-day moving average of the stock market against the bond market and the gold market. These two markets are less volitile in general, especially that we only have monthly data, and a monthly basis moving average would entail a good long-term trend. We also will be applying a typical 20-day Bollinger Band. The syntax for engineering the Bollinger Band is from Sajid Lhessani's Medium blog post: [Algorithmic trading: Full Python application of Bollinger Bands](https://medium.com/@lhessani.sa/beat-the-market-using-algorithmic-trading-and-bollinger-band-python-application-9a59d0b34451).

```python
## Features for monthly moving average
# SPX EMA13 for relatively short term momentum, capturing weekly momentum
df['SPX EMA13'] = df['SPX Close'].ewm(span=13).mean()

# SPX EMA 26 for relatively short term momentum, capturing monthly momentum
df['SPX EMA26'] = df['SPX Close'].ewm(span=26).mean()

# SPX/US10Y Relative Strength Line, 20-day simple moving average to capture general trend 
df['SPX/US10Y RSL MA20'] = df['SPX/US10Y RSL'].rolling(window=20).mean()

# SPX/GOLD Relative Strength Line, 20-day simple moving average to capture general trend 
df['SPX/GOLD RSL MA20'] = df['SPX/GOLD RSL'].rolling(window=20).mean()

# Applying Bollinger for monthly moving average
df['Bollinger 20 Middle Band'] = df['SPX Close'].rolling(window=20).mean()
df['Bollinger 20 Upper Band'] = df['Bollinger 20 Middle Band'] + 1.96*df['SPX Close'].rolling(window=20).std()
df['Bollinger 20 Lower Band'] = df['Bollinger 20 Middle Band'] - 1.96*df['SPX Close'].rolling(window=20).std()
```

Further engineering the quarterly data, here we include a EMA50 for the stock market again. We slightly modified the Bollinger Band syntax to also capture quarterly volitility. We also include a `GOLD/US10Y` pair for 60-day moving average RSL to find the longer-term relationship between the gold market and the bond market. As the gold market and the dollar index are also moderately correlated, we will also have a 60-day simple moving average for a RSL between the two to capture the longer-term dynamic.

```python
## Features for quarterly moving average
# Applying quarterly moving average for SPX
df['SPX EMA50'] = df['SPX Close'].ewm(span=50).mean()
df['Bollinger 60 Middle Band'] = df['SPX Close'].rolling(window=60).mean()
df['Bollinger 60 Upper Band'] = df['Bollinger 60 Middle Band'] + 1.96*df['SPX Close'].rolling(window=60).std()
df['Bollinger 60 Lower Band'] = df['Bollinger 60 Middle Band'] - 1.96*df['SPX Close'].rolling(window=60).std()

# GOLD/US10Y Relative Strength Line, 60-day simple moving average to capture general trend 
df['GOLD/US10Y RSL 60'] = df['GOLD/US10Y RSL'].rolling(window=60).mean()

# GOLD/DXY Relative Strength Line, 60-day simple moving average to capture general trend 
df['GOLD/DXY RSL MA60'] = (df['GOLD Price'] / df['DXY Close']).rolling(window=60).mean()
```

Finally, just as we have the RSL between the markets on a daily basis, we will have a EMA200 for long-term relationship.

```python
## EMA on RSL
# SPX/DXY RSL
df['SPX/DXY RSL EMA 200'] = df['SPX/DXY RSL'].ewm(span = 200).mean()

# SPX/WTI RSL
df['SPX/WTI RSL EMA 200'] = df['SPX/WTI RSL'].ewm(span = 200).mean()

# SPX/US10Y RSL
df['SPX/US10Y RSL EMA 200'] = df['SPX/US10Y RSL'].ewm(span = 200).mean()

# SPX/GOLD RSL
df['SPX/GOLD RSL EMA 200'] = df['SPX/GOLD RSL'].ewm(span = 200).mean()

# GOLD/US10Y RSL
df['GOLD/US10Y RSL EMA 200'] = df['GOLD/US10Y RSL'].ewm(span = 200).mean()
```

As we are working with multiple moving average features, we would also be introducing new missing values. The Exponential Weighted Functions does not introduce missing values, but simple moving averages do, so we would have at least 59 days (we have features of 60-day moving average) from the beginning of the dataset, but those will be treated soon. Even though we plan to create 3 datasets of shorter timeframes(5, 10, and 20 trading days), we will create values of multiple timeframes in each new dataset, ensuring that on each data point, not only do we have values of 5, 10, and 20 trading days ago, but also further back to 60 trading days to represent values from a quarter ago. As we perform this operation, the missing values would also be removed.

## **Transform Time Series to a Structured Learning Problem**

For the purpose of our project, we will transform the time series data into a structured learning problem. That means, we will be using a set of features to predict a target. In this case, we will be using every signal/indicator to predict the S&P 500 value on any given day. But we cannot use the values on the same day to predict the market value, so we have to shift the data to the past. In other words, on index 2020-01-01, we would have the S&P 500 market value on that day, but the value for everything else (gold market, the Bolinger Bands, etc) would actually be that of 2020-12-01. In effect, we're using the values from a month ago to predict the market value on 2020-01-01.

We will be creating a new column called `SPX Today` as our target, as we will want to have old value for `SPX Close` as well. We will then shifting all the rest of the values by a number of trading days to capture that trend of change over time. We will not be shifting `SPX Today` for it is our target, and we will not be shifting `CPI Annual Rate` because it is one constant value over the year to represent the inflation environment of that year, and shifting this feature would actually have misrepresentation on the inflation environment.

*Note: The idea of transforming to structured learning problem comes from Machine Learning Mastery's [blog post](https://machinelearningmastery.com/convert-time-series-supervised-learning-problem-python/)*

```python
# Creating a new column that represents market value of the day
df['SPX Today'] = df['SPX Close']
```

```python
# Take an list of columns except SPX Today and CPI Annual Rate to shift
column_list = list(df.drop(['SPX Today','CPI Annual Rate'], axis = 1).columns)
```

```python
## Creating copies for datasets of three different timeframes
five_df = df.copy()
ten_df = df.copy()
twenty_df = df.copy()
```

Specifically, for the each dataframe, we will be shifting values of:

|Dataframe Name|Predicting|Days to Lag|
|:---:|:---:|:---:|
|five_df|5 trading days later|5,10,20,60|
|ten_df|10 trading days later|10,20,40,60|
|twenty_df|20 trading days later|20,40,50,60|

```python
## shifting for five_df
# For every column in the column list
for c in column_list:
    for n in (5,10,20,60):      
        # Shift the data to 5-day, 10-day, 20-day, and 60-day lags, and name the new column with column name and n day lag
        five_df[f'{c} {n} Day Lag'] = five_df[f'{c}'].shift(n)
    # Dropping the original columns so that we do not have the same day value left
    five_df.drop(f'{c}', axis = 1, inplace = True)

# Dropping newly introduced missing values as  a result of shifting
five_df.dropna(axis = 0, how = 'any', inplace = True)
```

```python
## shifting for ten_df
# For every column in the column list
for c in column_list:
    for n in (10,20,40,60):      
        # Shift the data to 10-day, 20-day, 40-day, and 60-day lags, and name the new column with column name and n day lag
        ten_df[f'{c} {n} Day Lag'] = ten_df[f'{c}'].shift(n)
    # Dropping the original columns so that we do not have the same day value left
    ten_df.drop(f'{c}', axis = 1, inplace = True)

# Dropping newly introduced missing values as  a result of shifting
ten_df.dropna(axis = 0, how = 'any', inplace = True)
```

```python
## shifting for twenty_df
# For every column in the column list
for c in column_list:
    for n in (20,40,50,60):      
        # Shift the data to 20-day, 40-day, 50-day, and 60-day lags, and name the new column with column name and n day lag
        twenty_df[f'{c} {n} Day Lag'] = twenty_df[f'{c}'].shift(n)
    # Dropping the original columns so that we do not have the same day value left
    twenty_df.drop(f'{c}', axis = 1, inplace = True)

# Dropping newly introduced missing values as  a result of shifting
twenty_df.dropna(axis = 0, how = 'any', inplace = True)
```
